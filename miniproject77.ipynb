## Visualizing Convergence with Different Learning Rates

///Objective: Run the 1D gradient descent from the mini-project with three different learning rates (e.g., 0.01, 0.1, 1.0) and plot the paths on the same graph to demonstrate the effect of learning rate.///

```python
import numpy as np
import matplotlib.pyplot as plt
```

///Define a function and derivative//

```python
def f(x):
    return x**2
def df(x):
    return 2*x
```

////Gradient Descent function//

```python
def gradient_descent(start_x,learning_rate,n_iterations):
    x=start_x
    path=[x]
    for i in range(n_iterations):
        grad=df(x)
        x=x-learning_rate*grad
        path.append(x)
    return np.array(path)

###parameters
start_x=10
n_iterations=20
learning_rates=[0.01,0.1,1.0]
colors=['r','g','b']
```

///Run Gradient Descent for each learning Rate//

```python
plt.figure(figsize=(10,6))
x_vals=np.linspace(-12,12,400)
plt.plot(x_vals,f(x_vals),'k--',label='f(x)=x^2')
for lr,color in zip(learning_rates,colors):
    path=gradient_descent(start_x,lr,n_iterations)
    plt.plot(path,f(path),'o-',color=color,label=f'lr={lr}')
plt.title("Effect of Learning Rate on Gradient Descent (1D)")
plt.xlabel("x value")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
