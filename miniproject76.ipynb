## Gradient Descent for Simple Linear Regression

///Objective: Implement gradient descent from scratch for a single-feature linear regression model///

```python
import numpy as np
import matplotlib.pyplot as plt
```

///Genearte Data with noise:y=2x+5+noise

```python
np.random.seed(42)
x=np.random.rand(100,1)*10
noise=np.random.randn(100,1)*2
y=2*x+5+noise
```

///Initialize parameters//

```python
m=np.random.randn() ##slope
b=np.random.randn() #intercept
```

///set Hyperparameters//

```python
learning_rate=0.01
epochs=1000
```

///Gradient Descent Loop//

```python
n=len(x)
loss_history=[]
```

```python
for epochs in range(epochs):
    y_pred=m*x+b #predictions
    loss=np.mean((y_pred-y)**2) ##compute loss function(MSE)
    loss_history.append(loss)
    dm=(2/n)*np.sum((y_pred-y)*x)  ##compute gradient using formula
    db=(2/n)*np.sum(y_pred-y)
    m=m-learning_rate*dm   ## update parameters
    b=b-learning_rate*db
    if epochs % 100 == 0: #print debug info every 100 epochs
        print(f"Epoch {epochs:4d} | Loss: {loss:.4f} | m: {m:.4f} | b: {b:.4f}")
print(f"final slope(m):{m:.3f}")
print(f"Final intercept (b):{b:.3f}")
    




```

////visualiztion///

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.scatter(x,y,color='blue',label='Data')
plt.plot(x,m*x+b,color='red',label='best fit line')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Linear Regression fit")
plt.legend()

plt.subplot(1,2,2)
plt.plot(range(len(loss_history)),loss_history,color='purple')
plt.xlabel("Epoch")
plt.ylabel("Mse loss")
plt.title("Loss convergence over epochs")
plt.tight_layout()
plt.show()
```
