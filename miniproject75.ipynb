## 1D Gradient Descent on a Parabola

////Objective: Implement gradient descent to find the minimum of a simple 1D function.///

```python
import numpy as np
import matplotlib.pyplot as plt
```

////Define Function and Derivative:///

```python
def f(x):
    return x**2
def f_der(x):
    return 2*x

```

////Initialization: Set a random starting point x_start.///

```python
np.random.seed(42)
x_start=np.random.uniform(-10,10)
x=x_start
```

////Parameters: Set a learning_rate and epochs.///

```python
learning_rate=0.1
epochs=25
```

////Assign x and y value:

```python
x_value=[x]
y_value=[f(x)]
```

///Loop:///

```python
for i in range(epochs):
    grad=f_der(x)
    x=x-learning_rate*grad
    x_value.append(x)
    y_value.append(f(x))
```

////visualiztion://

```python
x_line = np.linspace(-10, 10, 200)
y_line = f(x_line)

plt.figure(figsize=(8, 6))
plt.plot(x_line, y_line, label='f(x) = x', color='blue')
plt.scatter(x_value, y_value, color='red', label='Gradient Descent Steps')
plt.plot(x_value, y_value, color='red', linestyle='--')

# Annotate start and end points
plt.scatter(x_value[0], y_value[0], color='green', s=100, label='Start')
plt.scatter(x_value[-1], y_value[-1], color='black', s=100, label='Minimum')

plt.title('Gradient Descent Visualization for f(x) = x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
