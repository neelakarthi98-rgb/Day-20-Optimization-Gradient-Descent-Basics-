## Gradient Descent with Multiple Features

////Objective: Extend the linear regression project to handle two features (e.g., y = 2x1 + 3x2 + 5 + noise).//

```python
import numpy as np
import matplotlib.pyplot as plt
```

////create data://

```python
np.random.seed(42)
x1=np.random.randn(100,1)*10 ##values between 0 and 1
x2=np.random.randn(100,1)*5  ##values between 0 and 5
 
y=2*x1+3*x2+5+np.random.randn(100,1)*2
```

////combine column into one matrix(add a column of 1s for bias)

```python
x=np.hstack([x1,x2,np.ones((100,1))])
```

///Initialize parameters///

```python
params=np.zeros((3,1))
learning_rate=0.0005
n_iterations=1000
```

////Gradient Descent

```python
m=len(y)
for i in range(n_iterations):
    y_pred=x@params
    error=y_pred-y
    gradients=(2/m)*(x.T@error)
    params-=learning_rate*gradients
    if i%200==0:
        loss=np.mean(error**2)
        print(f"iteration {i}:Loss={loss:.4f}")
print("\n Learned parameters [m1,m2,b]:")
print(params.ravel())
```

//visualization

```python
y_pred_final = x @ params
plt.scatter(y, y_pred_final, color='blue')
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.title("Actual vs Predicted (Two-Feature Linear Regression)")
plt.grid(True)
plt.show()
```
